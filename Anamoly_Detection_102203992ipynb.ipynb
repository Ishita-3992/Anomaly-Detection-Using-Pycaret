```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# List of models to evaluate
models = [
    'iforest','abod', 'cluster', 'cof', 'iforest', 'histogram',
    'knn', 'lof', 'svm', 'pca', 'mcd', 'sod', 'sos'
]
#Storing the anomaly counts and parameters for each model
count = {}
param = {}
silhouette_scores = {}
for i in models:
    model = create_model(i)
    # Predict anomalies
    res = assign_model(model)

    # Counting  the number of anomalies (where 'Anomaly' column is 1)
    #All positive values signify anomalies exist and vice versa true
    counts = res[res['Anomaly'] == 1].shape[0]
    count[i]=counts
    # parameters of the model
    params = model.get_params()
    param[i] = params

    clean_data = res[res['Anomaly'] == 0].drop(columns='Anomaly')

    # KMeans with 3 clusters
    kmeans = KMeans(n_clusters=3, random_state=123)
    kmeans.fit(clean_data)

    # silhouette score for the clustering
    score = silhouette_score(clean_data, kmeans.labels_)
    silhouette_scores[i] = score

# Display the anomaly counts, model parameters, and silhouette scores for each model
for i in count:
    print(f"Model: {i}")
    print(f"Anomaly Count: {count[i]}")
    print(f"Parameters: {param[i]}")
    print(f"Silhouette Score: {silhouette_scores[i]}\n")
```
